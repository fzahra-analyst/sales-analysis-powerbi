📘 Notebook 04 — Feature Engineering

```python
import pandas as pd

df = pd.read_csv("../data/processed/sales_clean.csv")
df.head()
```

```python
df["profit_margin"] = df["profit"] / df["sales"]
```

```python
df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")
```

```python
df["year"] = df["order_date"].dt.year
```

```python
df["month"] = df["order_date"].dt.month
```

```python
df["month_name"] = df["order_date"].dt.month_name()
```

3️⃣ Customer Segments (تقسيم العملاء)

```python
customer_orders = df.groupby("customer_id")["order_id"].count()

df["customer_segment"] = df["customer_id"].map(
    lambda x: "VIP" if customer_orders[x] >= 10 else "Regular"
)
```

```python
df.head()
```

```python
# remove wrong margin if exists
df = df.drop(columns=["profit_margin"], errors="ignore")
```

```python
# recalculate correctly
df = df[df["sales"] > 0]
df["profit_margin"] = (df["profit"] / df["sales"]) * 100
```

```python
df = df[df["sales"] > 10]  # threshold منطقي
```

```python
df.to_csv("../data/processed/sales_clean.csv", index=False)
```
